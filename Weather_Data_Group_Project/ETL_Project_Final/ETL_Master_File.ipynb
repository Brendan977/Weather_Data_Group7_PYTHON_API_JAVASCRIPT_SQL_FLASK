```python
weather_one = Path("weather_data1.csv")
```

```python
weather_one_df = pd.read_csv(weather_one)
```

```python
weather_one_df.drop(weather_one_df.iloc[:, 2:7], inplace=True, axis=1)
```

```python
dict1 = weather_one_df.to_dict()
```

```python
norm1 = pd.json_normalize(dict1)
```

```python
norm1.drop(norm1.iloc[:, 0:15], inplace=True, axis=1)
```

```python
df = pd.DataFrame(df1)

def parse_dict_string(dict_str):
    
    parsed_dict = json.loads(dict_str.replace("'", "\""))
    
    
    result_dict = {}
    for key, values in parsed_dict.items():
        result_dict[key] = values

    return result_dict
```

```python
df['parsed_data_NYC'] = df['NYC'].apply(parse_dict_string)

row_index = 0
parsed_dict = df.at[row_index, 'parsed_data_NYC']
time_values = parsed_dict['time']
carbon_monoxide_values = parsed_dict['carbon_monoxide']
uv_index_values = parsed_dict['uv_index']

specific_index = 1
specific_time_value = time_values[specific_index]
specific_carbon_monoxide_value = carbon_monoxide_values[specific_index]
specific_uv_index_value = uv_index_values[specific_index]
```

```python
df = df.join(pd.json_normalize(df['parsed_data_NYC']))
df = df.rename(columns={'time': 'NYC Date', 'carbon_monoxide': 'NYC Carbon Monoxide', 'uv_index': 'NYC UV-Index'}, inplace=False)
```

```python
df['parsed_data_Beijing'] = df['Beijing'].apply(parse_dict_string)

row_index = 0
parsed_dict = df.at[row_index, 'parsed_data_Beijing']
time_values = parsed_dict['time']
carbon_monoxide_values = parsed_dict['carbon_monoxide']
uv_index_values = parsed_dict['uv_index']

specific_index = 1
specific_time_value = time_values[specific_index]
specific_carbon_monoxide_value = carbon_monoxide_values[specific_index]
specific_uv_index_value = uv_index_values[specific_index]
```

```python
df = df.join(pd.json_normalize(df['parsed_data_Beijing']))
df = df.rename(columns={'time': 'Beijing Date', 'carbon_monoxide': 'Beijing Carbon Monoxide', 'uv_index': 'Beijing UV-Index'}, inplace=False)
```

```python
df['parsed_data_London'] = df['London'].apply(parse_dict_string)

row_index = 0
parsed_dict = df.at[row_index, 'parsed_data_London']
time_values = parsed_dict['time']
carbon_monoxide_values = parsed_dict['carbon_monoxide']
uv_index_values = parsed_dict['uv_index']

specific_index = 1
specific_time_value = time_values[specific_index]
specific_carbon_monoxide_value = carbon_monoxide_values[specific_index]
specific_uv_index_value = uv_index_values[specific_index]
```

```python
df = df.join(pd.json_normalize(df['parsed_data_London']))
df = df.rename(columns={'time': 'London Date', 'carbon_monoxide': 'London Carbon Monoxide', 'uv_index': 'London UV-Index'}, inplace=False)
```

```python
df['parsed_data_Tokyo'] = df['Tokyo'].apply(parse_dict_string)

row_index = 0
parsed_dict = df.at[row_index, 'parsed_data_Tokyo']
time_values = parsed_dict['time']
carbon_monoxide_values = parsed_dict['carbon_monoxide']
uv_index_values = parsed_dict['uv_index']

specific_index = 1
specific_time_value = time_values[specific_index]
specific_carbon_monoxide_value = carbon_monoxide_values[specific_index]
specific_uv_index_value = uv_index_values[specific_index]
```

```python
df = df.join(pd.json_normalize(df['parsed_data_Tokyo']))
df = df.rename(columns={'time': 'Tokyo Date', 'carbon_monoxide': 'Tokyo Carbon Monoxide', 'uv_index': 'Tokyo UV-Index'}, inplace=False)
```

```python
df['parsed_data_MX'] = df['Mexico City'].apply(parse_dict_string)

row_index = 0
parsed_dict = df.at[row_index, 'parsed_data_MX']
time_values = parsed_dict['time']
carbon_monoxide_values = parsed_dict['carbon_monoxide']
uv_index_values = parsed_dict['uv_index']

specific_index = 1
specific_time_value = time_values[specific_index]
specific_carbon_monoxide_value = carbon_monoxide_values[specific_index]
specific_uv_index_value = uv_index_values[specific_index]
```

```python
df = df.join(pd.json_normalize(df['parsed_data_MX']))
df = df.rename(columns={'time': 'MXC Date', 'carbon_monoxide': 'MXC Carbon Monoxide', 'uv_index': 'MXC UV-Index'}, inplace=False)
```

```python
df = df.drop(columns=['NYC', 'Beijing', 'London', 'Tokyo', 'Mexico City', 'parsed_data_NYC', 'parsed_data_Beijing', 'parsed_data_London', 'parsed_data_Tokyo', 'parsed_data_MX'])
```

```python
df = df.apply(lambda x: x.explode()).reset_index(drop=True)

df.reset_index(drop=True, inplace=True)

df.columns = [col.split()[0] if 'Date' in col else col for col in df.columns]
```

```python
df['NYC'] = pd.to_datetime(df['NYC'])
df['Beijing'] = pd.to_datetime(df['Beijing'])
df['London'] = pd.to_datetime(df['London'])
df['Tokyo'] = pd.to_datetime(df['Tokyo'])
df['MXC'] = pd.to_datetime(df['MXC'])

df.head()
```
